```python
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
```

```python
data = pd.read_csv(r'GroceriesInitial.csv')
#print(data[(data["basket_value"] < 2.7)]['id'].count())
#print(data[(data["basket_value"] >= 2.7) & (data["basket_value"] < 5.8)]['id'].count())
#print(data[(data["basket_value"] >= 5.8)]['id'].count())
data['basket'] = ''
data.loc[data['basket_value'] <= 2.7, 'basket'] = 'low_value_basket'
data.loc[(data['basket_value'] > 2.7) & (data['basket_value'] <= 5.8), 'basket'] = 'medium_value_basket'
data.loc[data['basket_value'] > 5.8, 'basket'] = 'high_value_basket'

data = data.drop(columns=['id', 'basket_value', 'recency_days'])
data = data.fillna('')
transactionEncoder = TransactionEncoder()     
array = transactionEncoder.fit(data.values).transform(data.values)
array = np.delete(array, 0, 0)
array = np.delete(array, 0, 1)
transactionEncoder.columns_.pop(0)
d = pd.DataFrame(array, columns=transactionEncoder.columns_)
d = d[['low_value_basket', 'medium_value_basket', 'high_value_basket', 'citrus fruit','tropical fruit','whole milk','other vegetables','rolls/buns', 'chocolate', 'bottled water', 'yogurt', 'sausage', 'root vegetables', 'pastry', 'soda', 'cream']]
```

```python
from apyori import apriori
from mlxtend.frequent_patterns import apriori, association_rules
# Intializing the list
# transacts = []
# populating a list of transactions
# for i in range(0, len(d)): 
  # transacts.append([str(data.values[i,j]) for j in range(0, 16)])

frequent_itemsets = apriori(d, min_support=0.03, use_colnames=True)
association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6).sort_values(by=['confidence'], ascending=False)
```
